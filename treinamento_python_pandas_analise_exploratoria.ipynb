```python
df["Valor Venda"].sum() #Soma do valor de venda
```

```python
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criação da coluna custo
```

```python
df.head(1)
```

```python
round(df["Custo"].sum(),2)
```

```python
df["Lucro"] = df["Valor Venda"] - df["Custo"] #Criação da Coluna Lucro
```

```python
df.head(1)
```

```python
round(df["Lucro"].sum(),2) #Soma do lucro, arredondando em 2 casas
```

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days #Criação da coluna de Tempo_envio
```
